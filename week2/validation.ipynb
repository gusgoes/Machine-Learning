```python
from sklearn.metrics import mean_squared_error
```

```python
max_poly = 10
degrees = range(1, max_poly + 1)
test_scores = np.zeros(max_poly + 1)
train_scores = np.zeros(max_poly + 1)
error_train = np.zeros(max_poly + 1)
error_test = np.zeros(max_poly + 1)
```

```python
for i in degrees:
    if i != 0:
        X_train_poly1 = PolynomialFeatures(i).fit_transform(X_train)
        X_test_poly1 = PolynomialFeatures(i).fit_transform(X_test)
        modelpoly1 = LinearRegression()
        modelpoly1.fit(X_train_poly1, y_train)
        train_scores[i] = modelpoly1.score(X_train_poly1, y_train)
        test_scores[i] = modelpoly1.score(X_test_poly1, y_test)
        error_train[i] = mean_squared_error(y_train, modelpoly1.predict(X_train_poly1))
        error_test[i] = mean_squared_error(y_test, modelpoly1.predict(X_test_poly1))
```

```python
bestdegree = np.argmin(error_test[1:]) + 1
print("Best degree:", bestdegree)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(list(degrees), error_train[1:], marker='o', label='train (in-sample)')
plt.plot(list(degrees), error_test[1:], marker='o', label='test')
plt.axvline(x=bestdegree, ymin=0, ymax=0.5, color='g', label="min test error at d=%d"%bestdegree, alpha=0.3)
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error')
plt.title('Train vs Test Error')
plt.legend(loc='upper left')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.show()
```
