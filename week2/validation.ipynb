```python
max_poly = 10
degrees = range(1, max_poly + 1)
test_scores = np.zeros(max_poly + 1)
train_scores = np.zeros(max_poly + 1)
error_train = np.zeros(max_poly + 1)
error_test = np.zeros(max_poly + 1)
```

```python
for i in degrees:
    if i != 0:
        X_train_poly = PolynomialFeatures(i).fit_transform(X_train)
        X_test_poly = PolynomialFeatures(i).fit_transform(X_test)
        polymodel = LinearRegression()
        polymodel.fit(X_train_poly, y_train)
        prediction_on_training = polymodel.predict(X_train_poly)
        prediction_on_test = polymodel.predict(X_test_poly)
        error_train[i] = mean_squared_error(y_train, prediction_on_training)
        error_test[i] = mean_squared_error(y_test, prediction_on_test)
        train_scores[i] = polymodel.score(X_train_poly, y_train)
        test_scores[i] = polymodel.score(X_test_poly, y_test)
```

## Validation

```python
X_v_train, X_v_valid, y_v_train, y_v_valid = train_test_split(X_train, y_train)
X_v_train.shape
```

```python
error_train = np.empty(len(degrees))
error_valid = np.empty(len(degrees))
score_train = np.empty(len(degrees))
score_valid = np.empty(len(degrees))
print( f'The error_train is {error_train.mean()}')
print( f'The error_valid is {error_valid.mean()}')
print( f'The score_train is {score_train.mean()}')
print( f'The score_valid is {score_valid.mean()}')
```

```python
for d in degrees:
    idx = d - 1  # degrees start at 1, arrays are 0-indexed (length == len(degrees))
    X_c = PolynomialFeatures(d).fit_transform(X_v_train)
    X_c_val = PolynomialFeatures(d).fit_transform(X_v_valid)
    est = LinearRegression()
    est.fit(X_c, y_v_train)
    error_train[idx] = mean_squared_error(est.predict(X_c), y_v_train)
    error_valid[idx] = mean_squared_error(est.predict(X_c_val), y_v_valid)
    score_train[idx] = est.score(X_c, y_v_train)
    score_valid[idx] = est.score(X_c_val, y_v_valid)
```

```python
#Plot the results
plt.figure(figsize=(10, 6))
plt.plot(degrees, error_train, marker='o', label='Train Error')
plt.plot(degrees, error_valid, marker='o', label='Validation Error')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.title('Error vs. Polynomial Degree')
plt.grid()
plt.show()
```
