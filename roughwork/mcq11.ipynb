## 📊 Grading Section

```python
# ANSWER KEY AND GRADING
correct_answers = {
    1: "B",  # Linear regression equation
    2: "C",  # Strong correlation
    3: "B",  # Train-test split
    4: "B",  # Overfitting
    5: "B",  # Logistic regression purpose
    6: "B",  # Cross-validation
    7: "B",  # Feature scaling importance
    8: "B",  # Learning rate
    9: "B",  # MSE purpose
    10: "C" # When to use test set
}

def grade_exam():
    score = 0
    total_questions = len(correct_answers)
    
    print("🎯 MCQ11 EXAM RESULTS")
    print("=" * 60)
    
    for question_num in range(1, total_questions + 1):
        student_answer = student_answers.get(question_num, "")
        correct_answer = correct_answers[question_num]
        
        if student_answer == correct_answer:
            print(f"✅ Question {question_num}: {student_answer} (Correct)")
            score += 1
        else:
            print(f"❌ Question {question_num}: {student_answer} (Correct answer: {correct_answer})")
    
    percentage = (score / total_questions) * 100
    
    print("\n" + "=" * 60)
    print(f"📈 FINAL SCORE: {score}/{total_questions} ({percentage:.1f}%)")
    
    # Grade classification
    if percentage >= 90:
        grade = "A+ (Excellent! 🌟)"
    elif percentage >= 80:
        grade = "A (Very Good! 🎉)"
    elif percentage >= 70:
        grade = "B (Good! 👍)"
    elif percentage >= 60:
        grade = "C (Pass 📚)"
    else:
        grade = "F (Review required 📖)"
    
    print(f"🏆 GRADE: {grade}")
    print("=" * 60)
    
    # Learning recommendations
    if percentage < 70:
        print("\n📚 STUDY RECOMMENDATIONS:")
        if score < 5:
            print("• Review Linear Regression fundamentals (Week 1)")
            print("• Practice train-test split concepts (Week 2)")
        if score < 7:
            print("• Study Logistic Regression applications (Week 3)")
            print("• Understand Cross-Validation techniques (Week 2)")
        if score < 9:
            print("• Review Gradient Descent concepts (Week 5)")
            print("• Practice Feature Scaling scenarios (Week 4)")
    
    return score, percentage

# Execute grading
if len(student_answers) == 10:
    final_score, final_percentage = grade_exam()
else:
    print(f"⚠️ Please answer all questions. Current: {len(student_answers)}/10")
```
