```python
# QUESTÃO 1 - ERRO COMUM: Data Leakage no Scaling
codigo_q1 = '''# Código do estagiário João:
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Carregando dados
bc = load_breast_cancer()
X, y = bc.data, bc.target

# Normalizando ANTES do split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividindo dados
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Treinando modelo
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print(f"Accuracy: {knn.score(X_test, y_test):.3f}")'''

mostrar_codigo_questao(1, "Data Leakage Clássico", codigo_q1,
    "Qual é o PRINCIPAL problema neste código?",
    {
        "A": "KNN com K=5 não é adequado para este dataset",
        "B": "StandardScaler foi aplicado antes do train_test_split (data leakage)",
        "C": "test_size=0.2 é muito pequeno, deveria ser 0.3",
        "D": "Faltou validação cruzada para escolher o melhor K"
    },
    "O scaler 'viu' dados do futuro que não deveria conhecer"
)

print("SUA RESPOSTA: ___")
```

```python
# QUESTÃO 2 - ERRO COMUM: Cross-validation incorreto
codigo_q2 = '''# Código da estagiária Maria:
from sklearn.model_selection import KFold
import numpy as np

# Setup para cross-validation
kf = KFold(n_splits=5, shuffle=False)
k_values = range(1, 21)
scores = []

for k in k_values:
    fold_scores = []
    
    for train_idx, val_idx in kf.split(X_test):  # USANDO X_test!
        X_fold_train, X_fold_val = X_test[train_idx], X_test[val_idx]
        y_fold_train, y_fold_val = y_test[train_idx], y_test[val_idx]
        
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(X_fold_train, y_fold_train)
        score = model.score(X_fold_val, y_fold_val)
        fold_scores.append(score)
    
    scores.append(np.mean(fold_scores))

best_k = k_values[np.argmax(scores)]
print(f"Melhor K: {best_k}")'''

mostrar_codigo_questao(2, "Cross-Validation Errado", codigo_q2,
    "Qual erro FATAL Maria cometeu?",
    {
        "A": "Usou shuffle=False no KFold, deveria ser True",
        "B": "Está fazendo cross-validation no conjunto de TESTE",
        "C": "Deveria usar StratifiedKFold para dados desequilibrados",
        "D": "n_splits=5 é muito pouco, deveria ser 10"
    },
    "Nunca toque no conjunto de teste até a avaliação final!"
)

print("SUA RESPOSTA: ___")
```

```python
# QUESTÃO 3 - ERRO COMUM: Pipeline mal implementado
codigo_q3 = '''# Código do estagiário Pedro:
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Criando pipeline
pipeline = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=7))

# Cross-validation com pipeline
for train_idx, val_idx in kf.split(X_train):
    X_fold_train, X_fold_val = X_train[train_idx], X_train[val_idx]
    y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]
    
    # Aplicando scaling manualmente MESMO tendo pipeline
    scaler = StandardScaler()
    X_fold_train_scaled = scaler.fit_transform(X_fold_train)
    X_fold_val_scaled = scaler.transform(X_fold_val)
    
    # Treinando pipeline com dados já escalados
    pipeline.fit(X_fold_train_scaled, y_fold_train)
    score = pipeline.score(X_fold_val_scaled, y_fold_val)
    print(f"Score: {score:.3f}")'''

mostrar_codigo_questao(3, "Pipeline Duplicado", codigo_q3,
    "O que Pedro fez de errado?",
    {
        "A": "Deveria usar StandardScaler() sem parâmetros",
        "B": "Está aplicando scaling duplo: manual + pipeline",
        "C": "Pipeline deveria usar MinMaxScaler ao invés de StandardScaler",
        "D": "n_neighbors=7 não é adequado para este caso"
    },
    "Pipeline já faz scaling internamente!"
)

print("SUA RESPOSTA: ___")
```

```python
# QUESTÃO 10 - ERRO COMUM: Implementação de Pipeline manual incorreta
codigo_q10 = '''# Implementação manual de pipeline para "entender melhor":
def meu_pipeline_manual(X_train, X_test, y_train, y_test):
    # Scaling
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.fit_transform(X_test)  # ERRO SUTIL!
    
    # Modelo
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_scaled, y_train)
    
    # Avaliação
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    
    return train_score, test_score

train_acc, test_acc = meu_pipeline_manual(X_train, X_test, y_train, y_test)
print(f"Train: {train_acc:.3f}, Test: {test_acc:.3f}")'''

mostrar_codigo_questao(10, "Pipeline Manual Bugado", codigo_q10,
    "Qual é o erro CRÍTICO na função meu_pipeline_manual?",
    {
        "A": "Deveria usar MinMaxScaler ao invés de StandardScaler",
        "B": "fit_transform no teste - deveria ser apenas transform",
        "C": "KNN deveria ter K=3 para melhor performance",
        "D": "Função deveria retornar o modelo treinado também"
    },
    "Scaler só aprende com treino, aplica no teste!"
)

print("SUA RESPOSTA: ___")
```

```python
def calcular_pontuacao_debugging():
    print("🐍 CORREÇÃO SIMULADO 4.0 - DEBUGGING & ESTRUTURAL")
    print("🔧 Vamos avaliar suas habilidades de debugging!")
    print()
    
    # Digite suas 10 respostas
    suas_respostas = input("Digite suas 10 respostas (ex: ABCDABCDAB): ").upper()
    
    gabarito_debugging = "BBBBBBBCBB"
    
    if len(suas_respostas) != 10:
        print("❌ Erro: Digite exatamente 10 respostas!")
        return
    
    acertos = 0
    bugs_encontrados = []
    bugs_perdidos = []
    
    print("\n🔍 ANÁLISE DE DEBUGGING - CODE REVIEW:")
    print("="*50)
    
    # Mapeamento de questões para tipos de bugs
    tipos_bugs = {
        1: "Data Leakage",
        2: "Cross-Validation Misuse", 
        3: "Pipeline Duplication",
        4: "Metrics Misinterpretation",
        5: "Overfitting Blindness",
        6: "Reproducibility Issues",
        7: "NumPy Syntax Error",
        8: "Model Comparison Flaw",
        9: "Test Set Contamination",
        10: "Transform vs Fit_Transform"
    }
    
    for i in range(10):
        questao = i + 1
        sua_resp = suas_respostas[i]
        correta = gabarito_debugging[i]
        bug_tipo = tipos_bugs[questao]
        
        if sua_resp == correta:
            print(f"🐛✅ Q{questao} ({bug_tipo}): BUG ENCONTRADO! 🔍")
            acertos += 1
            bugs_encontrados.append(bug_tipo)
        else:
            print(f"🐛❌ Q{questao} ({bug_tipo}): Bug não detectado ({sua_resp}→{correta})")
            bugs_perdidos.append(bug_tipo)
    
    pontuacao = (acertos / 10) * 100
    
    print("\n" + "="*50)
    print(f"🎯 DEBUGGING SCORE: {acertos}/10 ({pontuacao:.0f}%)")
    
    # Avaliação por nível de debugging
    if acertos >= 9:
        print("🔧 SENIOR ML ENGINEER - DEBUG MASTER! 🔧")
        print("🏆 Você identificou quase todos os bugs críticos!")
        print("💼 Pronto para fazer code review de projetos ML!")
    elif acertos >= 7:
        print("🛠️ ML PRACTITIONER - BOM DEBUGGING! 🛠️")
        print("🎯 Habilidades sólidas de identificação de problemas!")
        print("📈 Continue praticando casos mais sutis!")
    elif acertos >= 5:
        print("📚 INTERMEDIATE - CONTINUE PRATICANDO! 📚")
        print("🔍 Detectou bugs básicos, foque nos mais sutis!")
        print("💡 Revise as áreas onde perdeu bugs!")
    elif acertos >= 3:
        print("🔄 JUNIOR - REVISE IMPLEMENTAÇÕES! 🔄")
        print("📖 Precisa praticar mais debugging de código ML!")
        print("🎯 Foque nos erros mais comuns primeiro!")
    else:
        print("📝 TRAINEE - ESTUDE FUNDAMENTOS! 📝")
        print("🚨 Muitos bugs críticos passaram despercebidos!")
        print("🔄 Volte aos conceitos básicos antes de debugging!")
    
    # Feedback específico por categoria
    if bugs_encontrados:
        print("\n🔍 BUGS DETECTADOS COM SUCESSO:")
        for bug in bugs_encontrados:
            print(f"✅ {bug}")
    
    if bugs_perdidos:
        print("\n🚨 BUGS QUE PASSARAM DESPERCEBIDOS:")
        for bug in bugs_perdidos:
            print(f"❌ {bug} - REVISAR!")
    
    # Dicas de melhoria
    print("\n💡 DICAS PARA MELHORAR DEBUGGING:")
    if acertos < 7:
        print("🔧 Pratique lendo código linha por linha")
        print("🎯 Foque em data leakage - erro mais comum")
        print("📚 Estude boas práticas de ML pipelines")
    
    print("\n🎓 PARABÉNS POR COMPLETAR O SIMULADO DE DEBUGGING!")
    print("🐍 Agora você conhece os erros mais comuns em ML!")

# Execute para ver seu nível de debugging
calcular_pontuacao_debugging()
```
