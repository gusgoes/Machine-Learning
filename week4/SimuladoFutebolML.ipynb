## 🎯 **ANÁLISE DO SEU DESEMPENHO**

Execute a célula abaixo, coloque suas respostas e veja sua pontuação!

```python
def calcular_pontuacao():
    # COLOQUE SUAS RESPOSTAS AQUI (exemplo: "ABCDABCDAB")
    suas_respostas = input("Digite suas 10 respostas (exemplo: ABCDABCDAB): ").upper()
    
    gabarito_correto = "BCBBBBBABB"
    
    if len(suas_respostas) != 10:
        print("❌ Erro: Digite exatamente 10 respostas!")
        return
    
    acertos = 0
    print("\n📊 ANÁLISE DETALHADA:")
    print("="*40)
    
    for i in range(10):
        questao = i + 1
        sua_resp = suas_respostas[i]
        correta = gabarito_correto[i]
        
        if sua_resp == correta:
            print(f"✅ Questão {questao}: {sua_resp} - CORRETO!")
            acertos += 1
        else:
            print(f"❌ Questão {questao}: {sua_resp} (Correto: {correta}) - ERRO")
    
    pontuacao = (acertos / 10) * 100
    
    print("\n" + "="*40)
    print(f"🎯 RESULTADO FINAL: {acertos}/10 ({pontuacao:.0f}%)")
    
    if acertos >= 9:
        print("🥇 EXPERT! Você domina Machine Learning!")
    elif acertos >= 7:
        print("🥈 AVANÇADO! Excelente conhecimento!")
    elif acertos >= 5:
        print("🥉 INTERMEDIÁRIO! Continue praticando!")
    else:
        print("📚 INICIANTE! Revise os conceitos fundamentais!")
    
    # Feedback específico por área
    print("\n💡 ÁREAS DE FOCO:")
    if suas_respostas[0] != 'B':
        print("- Revise Feature Scaling e seus impactos")
    if suas_respostas[1] != 'C' or suas_respostas[9] != 'C':
        print("- Estude metodologia de Train/Test/Validation")
    if suas_respostas[2] != 'B':
        print("- Aprofunde Cross-Validation")
    if suas_respostas[4] != 'B':
        print("- Entenda Overfitting vs Underfitting")

# Execute e digite suas respostas
calcular_pontuacao()
```
