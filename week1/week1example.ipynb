```python
import numpy as np
from sklearn.linear_model import LinearRegression
x=np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y=np.dot(x, np.array([1, 2])) + 3
# y=1 * x0 + 2 * x1 + 3
reg = LinearRegression().fit(x, y)
reg.score(x, y)

# Output:
# 0.9999999999999998
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(np.array([[3, 5]]))
```
